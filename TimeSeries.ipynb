```python
# Generate a registration number for each student for identification
def getRegNo():
    studentIndice = []
    for i in range(len(makData['Index No'])):
        registrationNo = "23/U/"+ str(i+1).zfill(3)
        studentIndice.append(registrationNo)
    return list(studentIndice)
```

```python
studentData = pd.DataFrame(data=getRegNo(),
                           columns=["Reg No"],
                           index=[x for x in range(len(makData["NAME"]))])
```

```python
studentData['NAME'] = makData['NAME']
```

```python
studentData['COURSE'] = makData['COURSE NAME']
```
