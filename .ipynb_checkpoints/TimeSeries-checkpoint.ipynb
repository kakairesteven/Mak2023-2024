```python
import pandas as pd
import numpy as np
```

```python
makData = pd.read_csv("preprocessed/finalDfV2.csv",
            index_col=0
           )
```

```python
makData
```

```python
# Generate a registration number for each student for identification
def getRegNo():
    studentIndice = []
    for i in range(len(makData['Index No'])):
        registrationNo = "23/U/"+ str(i+1).zfill(3) 
        studentIndice.append(registrationNo)
    return list(studentIndice)
```

```python
studentData = pd.DataFrame(data=getRegNo(),
                           columns=["Reg No"],
                           index=[x for x in range(len(makData["NAME"]))])
```

```python
studentData['NAME'] = makData['NAME']
```

```python
studentData['COURSE'] = makData['COURSE NAME']
```

```python
# Tuition paid over a period of a baseline frequency of 25 minutes
dates = pd.date_range(
#                       start="2023-08-01 00:00:00",
                      start=pd.to_datetime("2023-08-01 00:10:00").tz_localize("Africa/Kampala"),
#                       end="2023-12-24 23:59:00",
#                       end=pd.to_datetime("2023-12-24 23:59:00").tz_localize("Africa/Kampala"),
#                       tz="Africa/Kampala",
                      normalize=True,
                      name="paymentDates",
                      inclusive="both",
                     freq="D",
                     periods=146
                     )
```

```python
dates
```

```python
# Tuition payment dataframe
fndf = pd.DataFrame(
                    {"Reg No": 
    [random.choice(list(studentData['Reg No'][:])) for x in range(len(dates))],
                        "Tuition": 
    [random.randint(1, 6000) * 1000 for x in range(len(dates))]

                    },
                     index=[random.choice(dates) for x in range(len(dates))]
                   )
```
