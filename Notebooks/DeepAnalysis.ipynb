```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
frame = pd.read_csv("../Data/Private.csv",
           # index_col=0,
                   )
```

```python
frame
```

```python
frame.dropna(how="any")
```

```python
frame
```

```python
frame = frame.astype({"UACE YEAR": "int64",
             "CODE": "int64",
             })
```

```python
frame
```

```python
frame.dtypes
```

```python
plt.figure(figsize=(20,20));
frame.groupby(["UACE YEAR"])["INDEX NO"].count(   
).head(20).plot(
    kind="bar",
    rot='vertical',
    title="Students admitted by Year",
    x="Year",
    y="Students",
);
plt.savefig("studentsAdmitted1995-2016.png");
```

```python
plt.figure(figsize=(15,15));
frame.groupby(["UACE YEAR"])["INDEX NO"].count(   
).tail(5).plot(
    kind="bar",
    rot='vertical',
    title="Students admitted by Year",
    x="Year",
    y="Students",
);
plt.savefig("studentsAdmitted2016-2023.png");
```
