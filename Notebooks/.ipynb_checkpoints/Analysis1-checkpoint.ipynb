```python
# Sort by Total weight
df.sort_values(by=['TOTAL WEIGHT'],
                    ignore_index=True,
                    ascending=True,
                    kind='mergesort',
                   )

def getSchoolIndex(StudentIndex):
    schoolIndex = StudentIndex[:5]
    return schoolIndex
```
